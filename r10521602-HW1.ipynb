#### 0. import essential modules

```python

import pandas as pd
```

#### 1. Import CSV data file, and make column 'date' datetime format (5% + 5%)

```python
fn_data = './Rain_MD-Androy-Bekily_1d_19810101-20210530_CHIRPS.csv'
df_rain_1d = pd.read_csv(fn_data)
#df_rain_1d.head(5)
#print(type(df_rain_1d.iloc[0]['date']))
df_rain_1d['date'] = pd.to_datetime(df_rain_1d['date'], format = '%Y/%m/%d')
```

#### 2. Calculate rainfall totals for each month, and export to a CSV file named _Rain_MD-Androy-Bekily_1M_19810101-20210530_CHIRPS.csv_ (25% + 5%)

```python

df_rain_1M = df_rain_1d.resample('M', on = 'date').sum()
df_rain_1M.head(5)

fn_data_out = './Rain_MD-Androy-Bekily_1M_19810101-20210530_CHIRPS.csv'
df_rain_1M.to_csv(fn_data_out, float_format = '%.2f')
```

#### 3. Calculate maximum 1-day rainfall for each month, and export to a CSV file named _Rain_MD-Androy-Bekily_1d-1M-Max_19810101-20210530_CHIRPS.csv_ (25% + 5%)

```python
df_rain_1d_1M_Max = df_rain_1d.resample('M', on = 'date').max()
del df_rain_1d_1M_Max['date']
df_rain_1d_1M_Max.head(5)

fn_data_out2 = './Rain_MD-Androy-Bekily_1d-1M-Max_19810101-20210530_CHIRPS.csv'
df_rain_1d_1M_Max.to_csv(fn_data_out2, float_format = '%.2f')
```

#### 4. Calculate maximum 3-day rainfall for canledar month December for each year, and export to a CSV file named _Rain_MD-Androy-Bekily_3d-Dec-Max_19810101-20210530_CHIRPS.csv_ (25% + 5%)

```python
df_rain_3d = df_rain_1d.resample('3D', on = 'date').sum().reset_index()
df_rain_3d.head(5)

df_rain_3d_Max = df_rain_3d.resample('M', on = 'date').max()
df_rain_3d_Max

df_rain_3d_Dec_Max = df_rain_3d_Max[df_rain_3d_Max['date'].dt.month == 12]
del df_rain_3d_Dec_Max['date']
df_rain_3d_Dec_Max

fn_data_out3 = './Rain_MD-Androy-Bekily_3d-Dec-Max_19810101-20210530_CHIRPS.csv'
df_rain_3d_Dec_Max.to_csv(fn_data_out3, float_format = '%.2f')
```

### For your reference, CHIRPS (Climate Hazards Group InfraRed Precipitation with Station data) is a satellite-based precipitation product. It has provided quasi-global rainfall estimate for over 35 years. You can find more information about CHRIPS via this link: https://www.nature.com/articles/sdata201566
